```python
import requests

headers = {
    'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/137.0 Safari/537.36'
}

url = 'https://www.ambitionbox.com/list-of-companies?page=1'

webpage = requests.get(url, headers=headers)

print(webpage.status_code)
```

```python
print(webpage.text[:1000])
```

```python
from bs4 import BeautifulSoup

soup = BeautifulSoup(webpage.text, 'html.parser')

print(soup.title)
```

```python
for tag in soup.find_all('h2')[:5]:
    print(tag)
    print("-"*100)
```

```python
print(soup.find_all('h2')[1].parent)
```

```python
name = []
rating = []

for company in soup.find_all('h2', class_='companyCardWrapper__companyName'):
    name.append(company.text.strip())
   
print(name)
print(len(name))
```

```python
for tag in soup.find_all():
    if tag.text.strip() == "3.9":
        print(tag)
```
